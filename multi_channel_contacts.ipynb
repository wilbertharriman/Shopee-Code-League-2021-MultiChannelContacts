```
# Read dataset from JSON file
import json

with open('contacts.json') as json_file:
    itemData = json.load(json_file)
```

```
class Graph:
  def __init__(self, V):
    self.V = V
    self.adj = [[] for i in range(V)]

  def addEdge(self, u, v):
    if u == v:
      return
    self.adj[u].append(v)
    self.adj[v].append(u)
    
  def DFSUtil(self, temp, v, visited):
    visited[v] = True
    temp.append(v)
    for i in self.adj[v]:
      if visited[i] == False:
        temp = self.DFSUtil(temp, i, visited)
    return temp

  def connectedComponents(self):
    visited = []
    cc = []
    for i in range(self.V):
      visited.append(False)
    
    for v in range(self.V):
      if visited[v] == False:
        temp = []
        cc.append(self.DFSUtil(temp, v ,visited))
    return cc
```

```
emails = dict()
orderIds = dict()
phoneNumbers = dict()

for item in itemData:
  if item['Email'] != '':
    try:
        emails[item['Email']] += list(item['Id'])
    except:
        emails[item['Email']] = list()
        emails[item['Email']].append(item['Id'])
  if item['OrderId'] != '':
    try:
        orderIds[item['OrderId']] += list(item['Id'])
    except:
        orderIds[item['OrderId']] = list()
        orderIds[item['OrderId']].append(item['Id'])
  if item['Phone'] != '':
    try:
        phoneNumbers[item['Phone']] += list(item['Id'])
    except:
        phoneNumbers[item['Phone']] = list()
        phoneNumbers[item['Phone']].append(item['Id'])
```

```
# Create graph representation of dataset
g = Graph(len(itemData))
for item in itemData:
  if (item['Email'] != ''):
    g.addEdge(emails[item['Email']][0], item['Id'])
  if (item['OrderId'] != ''):
    g.addEdge(orderIds[item['OrderId']][0], item['Id'])
  if (item['Phone'] != ''):
    g.addEdge(phoneNumbers[item['Phone']][0], item['Id'])
```

```
cc = g.connectedComponents()

ans = dict()
for L in cc:
  for i in L:
    ans[i] = L

result = []

for i in range(500000):
  ans[i] = sorted(ans[i])
  sum = 0
  text = ''

  for j in ans[i]:
    sum += itemData[j]['Contacts']
  text = '-'.join([str(k) for k in ans[i]])
  text += ', '
  text += str(sum)
  result.append(text)
```

```
# Create csv file
import pandas as pd

solution = pd.DataFrame({'ticket_id': [i for i in range(len(itemData))], 'ticket_trace/contact': result}, columns=['ticket_id', 'ticket_trace/contact'])
solution.to_csv("solution.csv",index=False)
```
